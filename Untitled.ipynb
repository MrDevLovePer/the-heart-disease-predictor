```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
for col in df.columns[:-1]:
    print(col , [df[col].unique()])
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.describe()
```

```python
X=df.drop(columns=['target'])
Y=df['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
[X.shape, X_train.shape, X_test.shape]
```

```python
[Y.shape, Y_train.shape, Y_test.shape]
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction=model.predict(X_train)
accu=accuracy_score(X_train_prediction, Y_train)
print(accu)
X_test_prediction=model.predict(X_test)
accu2=accuracy_score(X_test_prediction, Y_test)
print(accu2)
```

# Building Predictive System

```python
for i in range(200):
    tst=X_test.iloc[i].values
    type(tst)
    #tst=tst.reshape(1,-1)


    prd=model.predict([tst])
    print(X_test.iloc[i],'\n',prd)


    if(prd==[1]):
        print('The person has heart disease')
    else:
        print('The Person do not have heart disease')
```

```python
import pickle
# open a file, where you ant to store the data
file = open('model.sav', 'wb')

# dump information to that file
pickle.dump(model, file)
```

```python
# save the model to disk
filename = 'model2.sav'
pickle.dump(model, open(filename, 'wb'))
 
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, Y_test)
```
